```python
#Definição do produto
class tProdutos:
  codigo = 0
  nome = ''
  precoUnitario = 0.00
  qtdEstoque = 0

#Definição da Nota Fiscal
class tCupomFiscal:
  codigo = 0
  nome = ''
  precoUnitario = 0.00
  precoTotal = 0.00
  qtdCompra = 0

#Cadastrar produtos
def cadastrarProduto():
  cod = int(input("Digite o código: "))
  for i in range(len(produtos)):
    if cod == produtos[i].codigo:
      return -1

  produto = tProdutos()
  produto.codigo = cod
  produto.nome = input("Digite o nome: ")
  produto.qtdEstoque = int(input("Digite a quantidade de estoque: "))
  produto.precoUnitario = float(input("Digite o preço unitário: "))
  return produto

#Atualizar produtos
def atualizarProduto(produto, codigoProcurado):
  posicao = consultarProduto(produto, codigoProcurado)
  if posicao != -1:
    print("\nDigite a operação para atualizar")
    print("1 - Preço unitário do produto")
    print("2 - Quantidade do produto")
    print("3 - Quantidade e Preço unitário do produto")
#    print("4 - Nome do produto")
    op = int(input())

    if op == 1:
      precoUnitarioNovo = float(input("Digite o novo preço unitário do produto: "))
      produto[posicao].precoUnitario = precoUnitarioNovo
    else:
      if op == 2:
        quantidadeNova = int(input("Digite a nova quantidade para o produto: "))
        produto[posicao].qtdEstoque = quantidadeNova
      else:
        if op == 3:
          quantidadeNova = int(input("Digite a nova quantidade para o produto: "))
          precoUnitarioNovo = float(input("Digite o novo preço unitário do produto: "))
          produto[posicao].qtdEstoque = quantidadeNova
          produto[posicao].precoUnitario = precoUnitarioNovo
#        else:
#          if op == 4:
#            nomeNovo = input("Digite o novo nome do produto: ")
#            produto[posicao].nome = nomeNovo
  else:
    print("Produto não pode ser atualizado, pois não foi encontrado!")

#Registrar compra
def registrarCompra(produto, codigoProcurado):
  #Valida se existe o produto
  posicao = consultarProduto(produto, codigoProcurado)
  if posicao != -1:
    print("Nome do produto: " , produto[posicao].nome)
    qtdAComprar = int(input("Digite a quantidade de produtos para comprar: "))
    if qtdAComprar > produto[posicao].qtdEstoque:
      print("Estoque insuficiente! A quantidade disponível para compra é: ", produto[posicao].qtdEstoque)
      return -1
    else:
      #Compra do produto
      CFe = tCupomFiscal()
      CFe.codigo = produto[posicao].codigo
      CFe.nome = produto[posicao].nome
      CFe.qtdCompra = qtdAComprar
      CFe.precoUnitario = produto[posicao].precoUnitario
      CFe.precoTotal = (qtdAComprar * produto[posicao].precoUnitario)
      #Atualiza o valor do estoque
      produto[posicao].qtdEstoque -= qtdAComprar
      return CFe
  else:
    print("Produto não encontrado.")
    return -1

#Consultar produto
def consultarProduto(produto, codigoProcurado):
  for i in range (len(produto)):
    if produto[i].codigo == codigoProcurado:
      return i
  return -1

#Listar produtos no relatório e no programa principal
def listarProduto(produto):
  print("| {:>6} | {:<16} | {:>10} | {:>13.2f}  |".format(produto.codigo, produto.nome, produto.qtdEstoque, produto.precoUnitario))

#Relatório de produtos
def relatorioProdutos(produto):
  print("| Código | Nome             | Quantidade | Preço Unitário |")
  print("|---------------------------------------------------------|")
  for i in range(len(produto)):
    #print("{:>6} | {:<16} | {:>10} | {:>13}".format(produto[i].codigo, produto[i].nome, produto[i].qtdEstoque, produto[i].precoUnitario))
    listarProduto(produto[i])
    print("|---------------------------------------------------------|")

#Relatório de Cupom Fiscal (carrinho de compras)
def carrinhoDeCompras(compra):
  print("\nItens do cupom fiscal\n")
  print("| Código | Nome             | Quantidade | Preço Unitário | Preço Total |")
  print("|-----------------------------------------------------------------------|")
  for i in range(len(compra)):
    print("| {:>6} | {:<16} | {:>10} | {:>13.2f}  | {:>11.2f} |".format(compra[i].codigo, compra[i].nome, compra[i].qtdCompra, compra[i].precoUnitario, compra[i].precoTotal))
    print("|-----------------------------------------------------------------------|")

#Valor total do cupom fiscal (carrinho)
def valorTotal(compra):
  valorTotal = 0.00
  for i in range (len(compra)):
    valorTotal += compra[i].precoTotal
  print("|  {:>56}".format("Total |"), "{:>11.2f} |".format(valorTotal))

#Menu
def menu():
  print("\nDigite a opção desejada:")
  print("1 - Cadastrar produtos")
  print("2 - Atualizar produtos")
  print("3 - Registrar compra")
  print("4 - Consultar produtos")
  print("5 - Relatório de produtos")
  print("0 - Sair")
  op = int(input())
  return op

#Programa Principal
produtos = []
compras = []
while True:
  op = menu()
  if op == 0:
    break
  else:
    if op == 1:
      print("\nCadastro do produto")
      codProduto = cadastrarProduto()
      if codProduto == -1:
        print("Código do produto já está cadastrado.")
      else:
        produtos.append(codProduto)
    else:
      if op == 2:
        print("\nAtualizando o produto")
        codigoProcurado = int(input("Digite o código do produto: "))
        atualizarProduto(produtos, codigoProcurado)
      else:
        if op == 3:
          print("\nRealizando a compra")
          while True:
            codigoProcurado = int(input("Digite o código do produto: "))
            codigoProdutoComprado = registrarCompra(produtos, codigoProcurado) 
            if codigoProdutoComprado != -1:
              compras.append(codigoProdutoComprado)
              carrinhoDeCompras(compras)
              valorTotal(compras)
              continuarCompra = input("\nDeseja continuar comprando? Digite S - Sim / N - Não: ").lower()

              if continuarCompra == 'n':
                break
        else:
          if op == 4:
            print("\nConsultando produto")
            codigoProcurado = int(input("Digite o código do produto: "))
            posicao = consultarProduto(produtos, codigoProcurado)
            if posicao == -1:
              print("Produto não cadastrado.")
            else:
              print("| Código | Nome             | Quantidade | Preço Unitário |")
              print("|---------------------------------------------------------|")
              listarProduto(produtos[posicao])
          else:
            print("\nRelatório dos produtos em estoque")
            if op == 5:
              relatorioProdutos(produtos)
            else:
              print("Opção inválida")
```
