```python
#Definição do que um produto possui
class produtos:
  codigo = 0
  nome = ''
  precoUnitario = 0.00
  qtdEstoque = 0

#Montar Nota Fiscal
class cupomFiscal:
  codigo = 0
  nome = ''
  precoUnitario = 0.00
  precoTotal = 0.00
  qtdCompra = 0

#Cadastrar produtos
def cadastrarProduto():
  produto = produtos()
  produto.codigo = int(input("Digite o código do produto: "))
  produto.nome = input("Digite o nome do produto: ")
  produto.precoUnitario = float(input("Digite o preço unitário do produto: "))
  produto.qtdEstoque = int(input("Digite a quantidade de estoque do produto: "))
  return produto

#Atualizar produtos
def atualizarProduto(produto, codigoProcurado):
  posicao = consultarProduto(produto, codigoProcurado)
  if posicao != -1:
    precoUnitarioNovo = float(input("Digite o novo preço unitário do produto: "))
    quantidadeNova = int(input("Digite a nova quantidade para o produto: "))
    produto[posicao].precoUnitario = precoUnitarioNovo
    produto[posicao].qtdEstoque = quantidadeNova
  else:
    print("Produto não pode ser atualizado, pois não foi encontrado")

#Registrar compra
def registrarCompra(produto, codigoProcurado):
  #Valida se existe o produto
  posicao = consultarProduto(produto, codigoProcurado)
  if posicao != -1:
    qtdAComprar = int(input("Digite a quantidade de produtos para comprar: "))
    if qtdAComprar > produto[posicao].qtdEstoque:
      print("Estoque insuficiente! A quantidade disponível para compra é: ", produto[posicao].qtdEstoque)
      return -1
    else:
      #Compra do produto
      CFe = cupomFiscal()
      CFe.codigo = produto[posicao].codigo
      CFe.nome = produto[posicao].nome
      CFe.qtdCompra = qtdAComprar
      CFe.precoUnitario = produto[posicao].precoUnitario
      CFe.precoTotal = (qtdAComprar * produto[posicao].precoUnitario)
      #Atualiza o valor do estoque
      produto[posicao].qtdEstoque -= qtdAComprar
      return CFe
  else:
    print("Produto não encontrado.")
    return -1

#Consultar produto
def consultarProduto(produto, codigoProcurado):
  for i in range (len(produto)):
    if produto[i].codigo == codigoProcurado:
      return i
  return -1

#Relatório de produtos
def relatorioProdutos(produto):
  print("Código |  Nome             | Quantidade  | Preço Unitário")
  print("-------------------------------------------------------")
  for i in range(len(produto)):
    print('%5d  | %-30s             | %5d | %.2f'%(produto[i].codigo, produto[i].nome, produto[i].qtdEstoque, produto[i].precoUnitario))

#Relatório de cupom fiscal
def carrinhoDeCompras(compra):
  print("Código |  Nome             | Quantidade  | Preço Unitário | Preço Total")
  print("-------------------------------------------------------")
  for i in range(len(compra)):
    print('%5d  | %-30s             | %5d | %.2f | %.2f'%(compra[i].codigo, compra[i].nome, compra[i].qtdCompra, compra[i].precoUnitario, compra[i].precoTotal))

#Valor total do carrinho
def valorTotal(compra):
  valorTotal = 0.00
  for i in range (len(compra)):
    valorTotal += compra[i].precoTotal
  print("Total", valorTotal)
    
#Menu
def menu():
  print("1 - Cadastrar produtos")
  print("2 - Atualizar produtos")
  print("3 - Registrar compra")
  print("4 - Consultar produtos")
  print("5 - Relatório de produtos")
  print("0 - Sair")
  op = int(input())
  return op

produto = []
compras = []
while True:
  op = menu()
  if op == 0:
    break
  else:
    if op == 1:
      produto.append(cadastrarProduto())
    else:
      if op == 2:
        codigoProcurado = int(input("Digite o código do produto: "))
        atualizarProduto(produto, codigoProcurado)
      else:
        if op == 3:
          while True:
            codigoProcurado = int(input("Digite o código do produto: "))
            codigoProdutoComprado = registrarCompra(produto, codigoProcurado) 
            if codigoProdutoComprado != -1:
              compras.append(codigoProdutoComprado)
              carrinhoDeCompras(compras)#ListaCarrinho
              valorTotal(compras)#ValorTotalDaCompra
              continuarCompra = input("Deseja continuar comprando? Digite S - Sim / N - Não")
              if continuarCompra == 'N' or continuarCompra == 'n':
                break
        else:
          if op == 4:
            codigoProcurado = int(input("Digite o código do produto: "))
            posicao = consultarProduto(produto, codigoProcurado)
            if posicao == -1:
              print("Produto não cadastrado")
            else:
              print("Produto cadastrado.")
          else:
            if op == 5:
              relatorioProdutos(produto)
            else:
              print("Opção inválida")

#Validação pra código diferentes
```
